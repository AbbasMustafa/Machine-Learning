##### **Neural Networks: Representation**

###### **Non-linear Hypothesis**

As we know that for complex function representation (Non-linear Function) we need Complex hypothesis (Non-linear Hypothests), for that we need to add more features to create complex hypothesis, Say you wanted to create a hypothesis from two (2) features that included all the quadratic terms $g(\theta_0 + \theta_1 x_1^2 + \theta_2 x_1 x_2  + \theta_1 x_2^2)$

That gives us (3) features, The formula for calulating how many feautres we need to creat (r) polynomial Complex function $\frac{(n+r-1)!}{r!(n-1)}!$ (n) = number of feature for example $\frac{(2+2-1)!}{2!(2-1)!} = 3$

Example: let our training set be a collection of 50 x 50 pixel black-and-white photographs, and our goal will be to classify which ones are photos of cars. Our feature set size is then n = 2500 if we compare every pair of pixels.

> With quadratic features, our growth is $\frac{(2500+2-1)!}{2!(2500-1)!}=3,126,250$ which is clearly impracticale.

And here is come **Neural Network** an alternate way to perform machine learning when we have complex hypotheses with many features.

##### **Model Representation**

Let's examine how we will represent a hypothesis function using neural networks.

At a very simple level, neurons are basically computational units that take input (**dendrites**) as electrical input (called **"spikes"**) that are channeled to outputs (**axons**)

brain-neuron-a.gif

In our Mathematical model our **Dendrites** will input features $x_1 x_2 \dots x_n$ and our **axon** will be our output .

In neural networks, we use the same logistic function as in classification: $\frac{1}{1 + e^{-\theta^Tx}}$ In neural networks however we sometimes call it a sigmoid (logistic) **activation function**.

Our **"theta"** parameters are sometimes instead called "weights" in the neural networks model.

In this model our $x_0$ input node is sometimes called the **"bias unit."** It is always equal to 1.

This is represntation of **perceprtron** (Only one Neuron)

The first layer is called the **"input layer"** and the final layer the **"output layer,"** which gives the final value computed on the hypothesis.

We can have intermediate layers of nodes between the input and output layers called the **"hidden layer."**

###### **Notion**

$a^{(j)}_i$ = "activation" of unit $i$ in layer $j$ \\
$\theta^{(j)}$=matrix of weights controlling function mapping from layer $j$ to layer $j+1$

f we had one hidden layer, it would look visually something like:

$x_0x_1x_2x_3 \rightarrow a^{(2)}_1a^{(2)}_2a^{(2)}_3 \rightarrow h_\theta(x)$

The values for each of the **"activation"** nodes is obtained as follows:

$  a^{(2)}_1 = g(\theta^{(1)}_{10}x_0+\theta^{(1)}_{11}x_1+\theta^{(1)}_{12}x_2+\theta^{(1)}_{13}x_3) \\
a^{(2)}_2 = g(\theta^{(1)}_{20}x_0+\theta^{(1)}_{21}x_1+\theta^{(1)}_{22}x_2+\theta^{(1)}_{23}x_3) \\
a^{(2)}_3 = g(\theta^{(1)}_{30}x_0+\theta^{(1)}_{31}x_1+\theta^{(1)}_{32}x_2+\theta^{(1)}_{33}x_3) \\

h_{\theta}(x) = a^{(3)}_1 = g(\theta^{(1)}_{10}a^{(2)}_0+\theta^{(1)}_{11}a^{(2)}_1+\theta^{(1)}_{12}a^{(2)}_2+\theta^{(1)}_{13}a^{(2)}_3) \\
$

##### **Vectorization :Neural Network represeantion**

Let's Vectorize our neual netwerk representation.

First of all we have to vectorize $\Theta$ weighs.
Each layer gets its own matrix of weights, $\Theta^{(j)}$

The dimensions of these matrices of weights is determined as follows: If network has $s_j$ units/nodes in layer $j$ and $s_{j+1}$ units in layer $j+1$, then $\Theta_{(j)}$ will be of dimension $s_{j+1} × (s_j+1)$ Note: **+1** comes from bais feature $\theta_0$.

Example: layer 1 has 2 ($s_j$) input nodes and layer 2 has 4 $(s_{j+1})$ activation nodes. Dimension of $\Theta^{(1)}$ s going to be 4×3 where $s_j = 2$ and $s_{j+1} = 4$, so $s_{j+1} \times (s_j + 1) = 4 \times 3$



###### **Code**

```
import numpy as np
np.random.seed(10)

# Input Features also called a_1 (S_j)
a_1 = np.random.random((2,1))

# a_2 hidden Layer (2nd layer) (S_j+1) 
a_2_shape = 3

# theta Vector # (4 , 3) Dimension (+1) is for bais term
theta_1 = np.random.random((a_2_shape, a_1.shape[0] + 1 ))

def summation(a, theta):
  # add bais term to a_0
  bais = np.array([[1]])
  a = np.vstack((bais,a))

  z = theta.dot(a)
  return z

z_2 = summation(a_1, theta_1)
print(f"The input of 2nd layer abd output of 1sr layer (without Activation Fuction):\n {z_2}")
```

###### **Activation Function**

Now this time to add Activation Fucntion, We're going to define a new variable $z_k^{(j)}$ that encompasses the parameters inside our g function.

$z_k^{(j)} = \theta^{(j-1)}_{k,0}a_0^{(j-1)}+\theta^{(j-1)}_{k,1}a_1^{(j-1)}+\theta^{(j-1)}_{k,2}a_2^{(j-1)}+\theta^{(j-1)}_{k,3}a_3^{(j-1)}$ 

$ z^{(j)} = \Theta^{(j-1)}a^{(j-1)} $

$ a^{(2)}_1 = g(z^{(2)}_1) \\
a^{(2)}_2 = g(z^{(2)}_2) \\
a^{(2)}_3 = g(z^{(2)}_3) \\ $


###### **Code**

```
# Sigmod Function (Activation Function)

def activation_function(Z):
  return 1/(1+np.exp(-Z))

a_2 = activation_function(z_2)
print(f"g(a_2): \n {a_2}")
```

###### **Final Output**

Now it's time to final Output, which is one value in case of binary classifcation.

$h_{\theta}(x) = a^{(3)}_1 = g(\theta^{(1)}_{10}a^{(2)}_0+\theta^{(1)}_{11}a^{(2)}_1+\theta^{(1)}_{12}a^{(2)}_2+\theta^{(1)}_{13}a^{(2)}_3) \\
$

###### **Code**

```
# Output value is 1 if ew classifiy binary Value
a_3_shape = 1

# theta_2
theta_2 = np.random.random((a_3_shape, a_2.shape[0] + 1 ))

z_3 = summation(a_2, theta_2)
a_3 = activation_function(z_3)

print(f"Final Output: {a_3}")
```
